```python
# Group CPS Monthly Data URLs
urls = [
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/jan24pub.csv",
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/feb24pub.csv",
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/mar24pub.csv",
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/apr24pub.csv",
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/may24pub.csv",
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/jun24pub.csv",
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/jul24pub.csv",
        "https://www2.census.gov/programs-surveys/cps/datasets/2024/basic/aug24pub.csv"
       ]
```

```python
# Read each url as a dataframe
cps_2024_list = list(map(pd.read_csv, urls))
```

```python
def cps_aggregator(input_list, rule_list):
    cps_2024_combined = pd.DataFrame()
    for df in input_list:
        if df.columns.equals(rule_list.columns):
            cps_2024_combined = pd.concat([cps_2024_combined, df], ignore_index = True)
        else:
            # Find Missing Columns
            missing_columns = rule_list.columns.difference(df.columns) 
            # Create a dataframe out of those identified missing columns
            missing_data = pd.DataFrame(columns = missing_columns, index = df.index)
            # Creates df_with_missing pandas object where missing_data columns are appended horizontally to df
            df_with_missing = pd.concat([df, missing_data], axis = 1)
            df_with_missing.fillna(-1, inplace = True)
            # Appends current df with the missing columns entered to the cps_2024_combined dataframe
            cps_2024_combined = pd.concat([cps_2024_combined, df_with_missing], ignore_index = True)
    return cps_2024_combined
```

```python
cps_2024_full = cps_aggregator(cps_2024_list, cps_2024_dimensions)
```
