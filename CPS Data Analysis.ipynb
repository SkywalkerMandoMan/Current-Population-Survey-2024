```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Create copy of 2024 Current Population Survey Data
CPS_2024_copy = CPS_2024
```

```python
CPS_2024_copy['gtco'].value_counts()
```

```python
# Gender Count
CPS_2024['pesex'].value_counts()
```

```python
# Gender Mapping
gender_df = pd.DataFrame({'Gender': CPS_2024['pesex']})
gender_map = {1: 'Male', 2: 'Female', -1: 'Unknown'}
CPS_2024_copy['pesex'] = gender_df['Gender'].map(gender_map)
```

```python
CPS_2024_copy.head()
```

```python
# Create Gender Distribution Histogram
counts = CPS_2024_copy['pesex'].value_counts()

# Plot the histogram using bar plot for categorical data
plt.bar(counts.index, counts.values, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Histogram of Gender Categories')

# Show the plot
plt.show()
```
